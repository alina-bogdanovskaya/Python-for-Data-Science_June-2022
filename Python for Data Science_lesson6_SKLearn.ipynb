Импортируйте библиотеки pandas и numpy.

Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. Создайте датафреймы X и y из этих данных.

Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.

Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.


```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
data = boston["data"]
feature_names = boston["feature_names"]
target = boston["target"]
```

```python
data.shape
```

```python
feature_names
```

```python
target[:25]
```

```python
X = pd.DataFrame(data, columns=feature_names)
X.info()
```

```python
y = pd.DataFrame(target, columns=["price"])
y.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
compare_prices = pd.DataFrame({"y_test": y_test["price"], "y_pred": y_pred.flatten()})
compare_prices.head(20)
```

```python
lr_r2 = r2_score(y_test['price'], y_pred.flatten())
lr_r2
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, plot_roc_curve, roc_curve
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.value_counts('Class', normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
df.head(10)
```

```python
X = df.loc[ : , df.columns != 'Class']
y = df['Class']
```

```python
type(X), type(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
parameters = [{'n_estimators': [10, 15], 'max_features': np.arange(3, 5), 'max_depth': np.arange(4, 7)}]
my_clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100), 
                        param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
my_clf.fit(X_train, y_train)
```

```python
my_clf.best_params_
```

```python
best_clf = RandomForestClassifier(n_estimators=15, max_depth=6, max_features=3, random_state=100)
best_clf.fit(X_train, y_train)
```

```python
y_pred_proba_train = best_clf.predict_proba(X_train)[:, 1]
y_pred_proba_test = best_clf.predict_proba(X_test)[:, 1]
FPR_train, TPR_train, _ = roc_curve(y_train, y_pred_proba_train)
roc_auc_train = round(roc_auc_score(y_train, y_pred_proba_train), 5)
FPR_test, TPR_test, _ = roc_curve(y_test, y_pred_proba_test)
roc_auc_test = round(roc_auc_score(y_test, y_pred_proba_test), 5)
```

```python
plt.plot(FPR_train, TPR_train, label='train data, ROC_AUC='+str(roc_auc_train))
plt.plot(FPR_test, TPR_test, label='test data, ROC_AUC='+str(roc_auc_test))
plt.legend()
plt.show()
```
